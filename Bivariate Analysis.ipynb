# <font color='orange'> 1. Scatterplot (Numerical-Numerical) </font>

## Bivariate Analysis

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

## Multi Variate Analysis

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],palette="Set2")
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'],palette="husl")
```

# <font color='orange'>2. Bar Plot (Numerical-Categorical) </font>

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'],palette="Set2")
```

# <font color='orange'>3. Box Plot (Numerical-Categorical) </font>

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

# <font color='orange'>4. Distplot (Numerical-Categorical) </font>

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

orange line represents the probability of surviving and blue line represnts the probability of not surviving

# <font color='orange'>5. HeatMap (Categorical-Categorical) </font>

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

# <font color='orange'>6. ClusterMap (Categorical-Categorical) </font>

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

Shows closeness of different values
